## Exercise 2 - Filtering

Return the following subsets of the dataframe.

1. How many listings are there with a price less than 100? 


2. Find how many listings there are in just Brooklyn.


3. Find how many listings there are in Brooklyn with a price less than 100.


4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.


5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  The prices for all other listings should be the same as the were before. 


6. What % of the rooms are private, and what % of the rooms are shared.  
    * Hint, use `.value_counts()`


```python
# 1. How many listings are there with a price less than 100?
listings_under_100 = listings_df["price"] < 100

listings_df[listings_under_100].shape[0]
```

```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn.
df_bk = listings_df[listings_df["neighbourhood_group"] == "Brooklyn"]
df_bk.shape[0]
```

```python
# 3. Find how many listings there are in Brooklyn with a price less than 100.
bk_price_under_100 = df_bk["price"] < 100
df_bk[bk_price_under_100].shape[0]

```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.
# How many total are there that have those names

names = ["Michael", "David", "John", "Daniel"]
host_name_condition = listings_df["host_name"].isin(names)

listings_df[host_name_condition].shape[0]
```

```python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before. 
condition = listings_df["neighbourhood_group"] == "Williamsberg"
pricing = listings_df["price"]
listings_df["adjusted_price"] = np.where(condition, pricing + 100, pricing)
listings_df.head()
```

```python
# 6. What % of the rooms are private, and what % of the rooms are shared.  
shared_room = listings_df["room_type"] == "Shared room"
private_room = listings_df["room_type"] == "Private room"

listings_df_size = listings_df.shape[0]
shared_room_percentage = (listings_df[shared_room].shape[0] / listings_df_size) * 100
private_room_percentage = (listings_df[private_room].shape[0] / listings_df_size) * 100


print(f"Shared Room Percentage: {round(shared_room_percentage, 5)}%")
print(f"Private Room Percentage: {round(private_room_percentage, 5)}%")

```

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.
listings_df.groupby("neighbourhood_group")["id"].count()
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 
listings_df.groupby("neighbourhood_group")["price"].mean()

```

```python
# 2.5. Using `groupby`, find the mean price for each room_type. 
listings_df.groupby("room_type")["price"].mean()
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 
listings_df.groupby("neighbourhood_group")["price"].agg(["min", "max"])

```

```python
# 4. Using `groupby`, find the mean price for each room_type in each neighbourhood_group.
listings_df.groupby(["room_type", "neighbourhood_group"])["price"].mean()
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.

listings_df.groupby(["room_type", "neighbourhood_group"])["price"].agg(["min", "max", "mean", "std"])


```

```python
# 1. Who was won Album of the Year in 2016?
year = grammy_df["year"] == 2016
category = grammy_df["category"] == "Album of the Year"
winner = grammy_df["winner"] == True



won_album_2016 = grammy_df[year & category & winner]["workers"]
won_album_2016
```

```python
# 2. Who won Best Rap Album in 2009? 
year = grammy_df["year"] == 2009
category = grammy_df["category"] == "Best Rap Album"
winner = grammy_df["winner"] == True

best_rap_album_2009 = grammy_df[year & category & winner]["workers"]
best_rap_album_2009
```

```python
# 3. How many awards was Kendrick Lamar nomiated for, and how many did he win...?
artist = grammy_df["workers"] == "Kendrick Lamar"
winner = grammy_df["winner"] == True

kendrick_nominations = grammy_df[artist]
kendrick_nominations.shape[0]

kendrick_awards = grammy_df[artist & winner]
kendrick_awards.shape[0]
```
